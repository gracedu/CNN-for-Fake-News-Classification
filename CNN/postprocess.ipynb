```python
from sklearn.feature_extraction import text
import nltk

#lower case
false_negative["content"] = false_negative["content"].str.lower()
false_positive["content"] = false_positive["content"].str.lower()

#remove puctuation and stop word

#lemmatization
```

```python
stop = text.ENGLISH_STOP_WORDS
false_negative["content"] = false_negative["content"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
false_positive["content"] = false_positive["content"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
false_negative["content"] = false_negative["content"].str.replace('[^\w\s]','')
false_positive["content"] = false_positive["content"].str.replace('[^\w\s]','')
```

```python
import nltk as nltk
nltk.download('wordnet')

```

```python
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 

def lemma(sentence):
    tokens = sentence.split()
    lemmaed_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmaed_tokens)


false_negative["content"] = false_negative["content"].apply(lemma)
false_positive["content"] = false_positive["content"].apply(lemma)
```
