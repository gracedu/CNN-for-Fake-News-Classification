```python
from wordcloud.wordcloud import WordCloud
from wordcloud import STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 

fn = pd.read_csv("false_negative.csv")
fp = pd.read_csv("false_positive.csv")
```

```python
comment_words1 = '' 
comment_words2 = '' 
stopwords = set(STOPWORDS) 


for val in fn["content"]: 
    val = str(val) 
    tokens = val.split()   
    comment_words1 += " ".join(tokens)+" "
    
    
for val in fp["content"]: 
    val = str(val) 
    tokens = val.split() 
    comment_words2 += " ".join(tokens)+" "
```

```python
from PIL import Image
wordcloud1 = WordCloud().generate(comment_words1)
wordcloud2 = WordCloud().generate(comment_words2)

plt.imshow(wordcloud1) 
plt.axis("off")
plt.show
wordcloud1.to_file("false_negative.png")
```
